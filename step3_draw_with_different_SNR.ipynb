```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 
data_frame = pd.read_csv('output_data_single.csv')

# SNR
sorted_snrs = sorted(data_frame['SNR'].unique())
n_plots = 4  # SNR

# 
fig, axs = plt.subplots(nrows=5, ncols=1, figsize=(15, 25), sharex=True)

for i in range(5):
    ax = axs[i]
    # SNR
    snrs_to_plot = sorted_snrs[i*n_plots:(i+1)*n_plots]
    for snr in snrs_to_plot:
        # SNR
        sample_data = data_frame[data_frame['SNR'] == snr].iloc[0, :-2]  # Mod_TypeSNR
        ax.plot(sample_data, label=f'SNR {snr}')

    # x50
    ax.set_xticks(np.arange(0, len(sample_data), 50))
    ax.set_xticklabels(np.arange(0, len(sample_data), 50))

    ax.set_title(f'Waveforms for SNR Levels {snrs_to_plot}')
    ax.set_xlabel('Sample Index')
    ax.set_ylabel('Amplitude')
    ax.legend()
    ax.grid(True)

# 
plt.suptitle('Waveforms for Different SNR Levels of QAM64 Modulation')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 
data_frame = pd.read_csv('output_data_single.csv')

# SNR
sorted_snrs = sorted(data_frame['SNR'].unique())
types = sorted(data_frame['Mod_Type'].unique())

# 
n_rows = len(sorted_snrs)

# 
fig, axs = plt.subplots(nrows=n_rows, ncols=1, figsize=(15, 5 * n_rows), sharex=True)

for i, snr in enumerate(sorted_snrs):
    ax = axs[i]
    for mod_type in types:
        # SNRType
        sample_data = data_frame[(data_frame['SNR'] == snr) & (data_frame['Mod_Type'] == mod_type)].iloc[0, :-2]  # Mod_TypeSNR
        ax.plot(sample_data, label=f'{mod_type}')

    # x50
    ax.set_xticks(np.arange(0, len(sample_data), 50))
    ax.set_xticklabels(np.arange(0, len(sample_data), 50))

    ax.set_title(f'Waveforms for SNR Level {snr}')
    ax.set_xlabel('Sample Index')
    ax.set_ylabel('Amplitude')
    ax.legend()
    ax.grid(True)

# 
plt.suptitle('Waveforms for Different Modulation Types at Each SNR Level')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # 
plt.show()
```
