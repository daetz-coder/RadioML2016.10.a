```python
import tarfile
import os
from scipy.io import loadmat
import numpy as np

# 文件路径
file_path = 'RML2016.10a.tar.bz2'
extract_path = './data'  # 解压到当前目录下的 data 文件夹

# 确保解压目录存在
if not os.path.exists(extract_path):
    os.makedirs(extract_path)

# 解压文件
with tarfile.open(file_path, "r:bz2") as tar:
    tar.extractall(path=extract_path)  # 指定解压路径
    print("Files extracted:")
    print(tar.getnames())  # 打印解压的文件列表
```
